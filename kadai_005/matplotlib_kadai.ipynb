```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(2, 3, tight_layout=True)
# line graph
y0 = [1, 2, -4, 3, -1, 0]
axes[0,0].plot(y0, linewidth = 3.0)

# sine function
x1 = np.linspace(0, 10, 100)
y1 = 3 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1, linewidth = 1.5)
axes[0, 1].set(xlim = (0, 10), xticks = np.arange(0, 10),
       ylim = (0, 6), yticks = np.arange(0, 6))

# histogram
x2 = np.random.normal(25, 10, 3000)
plt.xlim(-10, 50)
axes[0, 2].hist(x2, bins = 50)

# scatter plot
np.random.seed(42)
x3 = 4 + np.random.normal(0, 2, 250)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 0].scatter(x3, y3)

plt.show()
```
